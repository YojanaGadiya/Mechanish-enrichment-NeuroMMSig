```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
df = pd.read_csv("Citric_Acid_Cycle.txt",sep='\t',header=None)
df.columns = ['Prt1', 'Relation', 'Prt2']
df
```

```python
df['Relation'].value_counts()
```

```python
relations = {'used-to-produce':'increase','catalysis-precedes':'decrease', 'controls-production-of':'increase',
             'reacts-with':'decrease'}
print(type(relations))
```

```python
for i in relations:
    df.replace(i,relations[i],inplace=True)
```

```python
new_df = df.loc[df['Relation'].isin(relations.values())]
new_df
```

<h3>Network

```python
DG = nx.DiGraph()
DG.add_nodes_from(new_df['Prt1'])
DG.add_nodes_from(new_df['Prt2'])
for _,row in new_df.iterrows():
    DG.add_edge(row['Prt1'], row['Prt2'], **{"Relation":row['Relation']})
```

```python
edges = [
    data["Relation"]
    for source,target,data in DG.edges(data=True)
    if "Relation" in data
]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
nx.draw(DG, with_labels=True, font_weight='bold')
plt.show()
```
